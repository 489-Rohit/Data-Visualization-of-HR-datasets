## Data Cleaning Part

```python
# Removing commas from Employee_Name column
df['Employee_Name'] = df['Employee_Name'].str.replace(',', '')
```

```python
df['Employee_Name']
```

```python
# filling null value of column ManagerID with moode
df['ManagerID'].fillna(df['ManagerID'].mode()[0], inplace = True)
```

```python
df['ManagerID'].isnull().sum()
```

```python
# Cleaning of DOB column
def subtract_years(date):
    if date.year > 2000:
        return date.replace(year=date.year-100)
    else:
        return date
    
    df["DOB"] = df["DOB"].apply(subtract_years).dt.strftime('%d-%m-%Y')
```

```python
df["DOB"] = pd.to_datetime(df["DOB"])
df["DOB"].sort_values(ascending= True)
```

```python
# replacing duplicate value of HispanicLatino column

replace_dict = {'yes': 'Yes', 'no':'No'}
df['HispanicLatino'] = df['HispanicLatino'].replace(replace_dict)
```

```python
df['HispanicLatino'].unique()
```

```python
# changing data type of DateofHire colum

df["DateofHire"] = pd.to_datetime(df["DateofHire"])
```

```python
df['DateofHire'].info()
```

```python
# Removing extra space from Production attributes in Department column
replace_production = {'Production       ': 'Production'}
df['Department'] = df['Department'].replace(replace_production)
```

```python
df['Department'].unique()
```

```python
#typecastin of Manager ID from float to int

df['ManagerID'] = df['ManagerID'].astype(int)
```

```python
df['ManagerID'].info()
```

```python
df['ManagerID'].unique()
```

```python
#changing dtype of lastperformace column to datetime

df['LastPerformanceReview_Date'] = pd.to_datetime(df['LastPerformanceReview_Date'])
```

```python
df['LastPerformanceReview_Date'].info()
```

## Data Visulization part

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly
```

```python
# Checking the employment status of Employess using piechart

employment_status_counts = df.groupby('EmploymentStatus').size()

plt.pie(employment_status_counts, labels=employment_status_counts.index, autopct='%1.1f%%')


plt.title('Employment Status of Employees')
plt.show()
```

```python
sns.boxplot(x='DeptID', y='Salary', data=df)
```

Got to know that the person fall under DeptID 4 getting more salary compare to others


```python

grouped = df.groupby('Position')['Salary'].mean()

grouped = grouped.sort_values(ascending=False)

fig, ax = plt.subplots(figsize=(8, 6))
ax.barh(grouped.index, grouped.values)

ax.set_xlabel('Salary')
ax.set_ylabel('Position')
ax.set_title('Average Salary by Position')

plt.show()
```

Got the insight about Salary with respect to the position


```python
plt.hist(df['PerformanceScore'], bins=30)

plt.xlabel('Performance')
plt.ylabel('Score')
plt.title('Performance Distribution')

plt.show()
```

Most of the employees fully meets the performace 

```python
Recruitment = df.groupby('RecruitmentSource').size()

plt.pie(Recruitment, labels=Recruitment.index, autopct='%1.1f%%')


plt.title('RecruitmentSource')
plt.show()
```

Most of the people recruited via Indeed & linkedIn

```python
grouped = df.groupby('TermReason')['Employee_Name'].count().reset_index(name='Count')
grouped = grouped.sort_values('Count', ascending=False)

fig, ax = plt.subplots(figsize=(10, 6))
sns.barplot(x='Count', y='TermReason', data=grouped, ax=ax)
ax.set_title('Termination Reasons')
ax.set_xlabel('Count')
ax.set_ylabel('Reason')

plt.show()
```

Most Termination reason is another position

```python
grouped = df.groupby('MaritalDesc')['Employee_Name'].count()

fig, ax = plt.subplots(figsize=(8, 6))
ax.pie(grouped.values, labels=grouped.index, autopct='%1.1f%%')

ax.set_title('Employees by Marital Status')

plt.show()
```

Got the insight that the most of the employess of this company are Single

```python
grouped = df.groupby('Sex')['Employee_Name'].count()

fig, ax = plt.subplots(figsize=(8, 6))
ax.bar(grouped.index, grouped.values)

ax.set_title('Employees by Sex')
ax.set_xlabel('Sex')
ax.set_ylabel('Number of Employees')

plt.show()
```
